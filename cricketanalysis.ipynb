```python
#importing necessary libraries
import pandas as pd
import json
```

1. Match Result


```python
with open('/content/sample_data/t20_wc_match_results.json')as f:
  data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({"scorecard" :"match_id" } ,axis = 1 , inplace = True)
df_match.head()
```

Creation of match ids dictionary that maps team names to a unique match id .

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
  key1 = row['team1']+ ' Vs ' + row['team2']
  key2 = row['team2']+ ' Vs ' + row['team1']

  match_ids_dict[key1] = row["match_id"]
  match_ids_dict[key2] = row["match_id"]

match_ids_dict
```

```python

df_match.to_csv('/content/csv_files/dim_match_summary.csv', index = False)
```

2. Batting summary

```python
with open('/content/sample_data/t20_wc_batting_summary.json')as f:
  data = json.load(f)

  all_records = []

  for rec in data:
    all_records.extend(rec['battingSummary'])

df_batting =pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x) >0 else "not_out")
df_batting.head(11)
```

```python
df_batting.drop(columns=["dismissal"], inplace = True)
df_batting.head(10)
```

Cleanup weird characters


```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x : x.replace('†' , ' '))
df_batting['batsmanName'] =df_batting['batsmanName'].apply(lambda x: x.replace('\xa0' , ' '))
df_batting.head(11)
```

```python
match_ids_dict["Namibia Vs Sri Lanka"]
```

```python
df_batting.shape
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv('/content/csv_files/fact_batting_summary.csv', index = False)
```

3. Bowling Summary

```python
with open('/content/sample_data/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```
